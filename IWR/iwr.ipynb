```python
import seaborn as sns

# Select relevant columns for correlation analysis
correlation_cols = ['temperature_celsius', 'precip_mm', 'wind_kph', 'humidity', 'air_quality_PM2.5']

# Calculate the correlation matrix
correlation_matrix = iwr[correlation_cols].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation M
plt.show()
```

```python
var = 'timezone'
data = pd.concat([iwr['temperature_fahrenheit'], iwr[var]], axis=1)
f, ax = plt.subplots(figsize=(12, 8))
fig = sns.boxplot(x=var, y="temperature_fahrenheit", data=data)
fig.axis(ymin=0, ymax=165);
```

```python
var = 'moon_phase'
data = pd.concat([iwr['air_quality_PM10'], iwr[var]], axis=1)
f, ax = plt.subplots(figsize=(20, 10))
fig = sns.boxplot(x=var, y="air_quality_PM10", data=data)
fig.axis(ymin=0, ymax=165);
plt.xticks(rotation=90);
```

```python
moon_phase = iwr['moon_phase']
celsius = iwr['temperature_celsius']

plt.figure(figsize=(8,6))
plt.scatter(moon_phase,celsius,alpha=0.5)
plt.title("Scatter plot of moon phase vs Temperature(celsius)")
plt.xlabel("Moon phase")
plt.ylabel("Temperature in celsius")
plt.xticks(rotation=45) 
plt.grid(True)
plt.show()
```

```python
var = "region"
plt.figure(figsize=(20, 10))
sns.catplot(x=var, kind="count", palette="ch:.25", height=8, aspect=2, data=iwr);
plt.xticks(rotation=90);
```
